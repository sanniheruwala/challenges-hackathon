```python
from __future__ import print_function
import argparse
import torch
from PIL import Image
from torchvision.transforms import ToTensor
import import_ipynb
import numpy as np
from multiprocessing.dummy import Pool as ThreadPool
```

```python
# Training settings
parser = argparse.ArgumentParser(description='PyTorch Super Res Example')
parser.add_argument('--input_image', type=str, default="dataset/BSDS300/images/test/296007.jpg", help='input image to use')
parser.add_argument('--model', type=str , default="/weights/model_epoch_1000.pth",help='model file to use')
parser.add_argument('--output_filename', type=str , default="out1.png", help='where to save the output image')
parser.add_argument('--cuda', action='store_true', help='use cuda')
opt = parser.parse_args('')
```

```python
print(opt)
img = Image.open(opt.input_image).convert('YCbCr')
y, cb, cr = img.split()
```

```python
model = torch.load(opt.model)
img_to_tensor = ToTensor()
input = img_to_tensor(y).view(1, -1, y.size[1], y.size[0])
```

```python
if opt.cuda:
    model = model.cuda()
    input = input.cuda()
```

```python
out = model(input)
out = out.cpu()
out_img_y = out[0].detach().numpy()
out_img_y *= 255.0
out_img_y = out_img_y.clip(0, 255)
out_img_y = Image.fromarray(np.uint8(out_img_y[0]), mode='L')
```

```python
out_img_cb = cb.resize(out_img_y.size, Image.BICUBIC)
out_img_cr = cr.resize(out_img_y.size, Image.BICUBIC)
out_img = Image.merge('YCbCr', [out_img_y, out_img_cb, out_img_cr]).convert('RGB')
```

```python
out_img.save(opt.output_filename)
print('output image saved to ', opt.output_filename)
```

```python
def mergePerFrame(x):
    parser = argparse.ArgumentParser(description='Super Res')
    parser.add_argument('--input_image', type=str, default="frames_360p_new/"+str(x)+".jpg", help='input image to use')
    parser.add_argument('--model', type=str , default="weights/model_epoch_1000.pth",help='model file to use')
    parser.add_argument('--output_filename', type=str , default="frames_output_360p_new/frame"+str(x)+".png", help='where to save the output image')
    parser.add_argument('--cuda', action='store_true', help='use cuda')
    opt = parser.parse_args('')
    print(opt)
    img = Image.open(opt.input_image).convert('YCbCr')
    y, cb, cr = img.split()
    model = torch.load(opt.model)
    img_to_tensor = ToTensor()
    input = img_to_tensor(y).view(1, -1, y.size[1], y.size[0])
    if opt.cuda:
        model = model.cuda()
        input = input.cuda()
    out = model(input)
    out = out.cpu()
    out_img_y = out[0].detach().numpy()
    out_img_y *= 255.0
    out_img_y = out_img_y.clip(0, 255)
    out_img_y = Image.fromarray(np.uint8(out_img_y[0]), mode='L')
    out_img_cb = cb.resize(out_img_y.size, Image.BICUBIC)
    out_img_cr = cr.resize(out_img_y.size, Image.BICUBIC)
    out_img = Image.merge('YCbCr', [out_img_y, out_img_cb, out_img_cr]).convert('RGB')
    out_img.save(opt.output_filename)
    print('output image saved to ', opt.output_filename)
    
def calculateParallel(x, threads=2):
    pool = ThreadPool(threads)
    results = pool.map(mergePerFrame, x)
    pool.close()
    pool.join()
    return results

    
x = range(0, length)
calculateParallel(x, 8)
```
